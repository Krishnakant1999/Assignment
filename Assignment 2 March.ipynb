Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating static, animated, and interactive plots. It is extensively used for generating high-quality graphs, charts, and figures for data analysis and presentation.

Line Plot: A basic plot that displays data points connected by straight lines. It is useful for visualizing trends and patterns in continuous data.

Scatter Plot: It represents individual data points as markers on a graph, where the position of each point corresponds to its values on the x and y axes. Scatter plots are useful for showing the relationship between two variables.

Bar Plot: This plot displays categorical data using rectangular bars, where the height or length of each bar represents the values associated with each category. Bar plots are suitable for comparing data across different categories.

Histogram: It is used to represent the distribution of a continuous variable by dividing the data into bins and displaying the number of data points that fall into each bin.

Pie Chart: This circular chart represents the proportions of different categories as slices of a pie. Pie charts are commonly used to display the composition or relative sizes of categorical data.

plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

Note: Also add title, xlabel, and ylabel to the plot

```python
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
colours=np.random.rand(50)
sizes = 1000*colours
plt.scatter(x,y,c=colours,s=sizes,alpha=0.8)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter data")
```

Q.3. The subplot() function is used to create a grid of subplots within a single figure. It allows you to plot multiple plots or visualizations in a grid layout, making it easier to compare and analyze different data or results.
import numpy as np

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of chart that represents categorical data with rectangular bars. It is commonly used to compare and display the values of different categories or groups. The length or height of each bar corresponds to the magnitude of the data being represented.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Profit of Companies")
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.title("Profit of Companies")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays a summary of the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum values of a dataset. Additionally, it can show any outliers present in the data.

Box plots are used to summarize and compare the distribution of different datasets or variables. They provide information about the central tendency, spread, skewness, and presence of outliers in the data. 

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a list of data for the box plot
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
