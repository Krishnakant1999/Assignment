Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a popular web framework for building web applications in Python. It is a lightweight framework that follows the microframework approach, meaning it provides only the essential features needed to build web applications and keeps the core simple and extensible.

Advantages:
Flexibility: Flask gives developers the freedom to structure their applications as they see fit. It does not enforce any specific coding patterns or project structures, enabling developers to choose the best approach for their application.

Built-in Development Server and Debugger: Flask comes with a built-in development server, making it easy to test and run applications during the development process. It also provides a powerful debugger that helps identify and fix issues in the code.

Large and Active Community: Flask has a large and active community of developers, which means there is extensive documentation, tutorials, and resources available. This community support makes it easier for developers to find solutions, get help, and collaborate with others.

Well-suited for Small to Medium-Sized Projects: Flask's lightweight and flexible nature make it a good choice for small to medium-sized projects where simplicity and speed are important. It allows developers to quickly prototype and build applications without unnecessary complexity.

It's worth noting that while Flask is a powerful and popular framework, it may not be the best choice for all types of projects. For large-scale applications or projects with complex requirements, other frameworks like Django might provide more out-of-the-box features and conventions.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or view handlers within the web application. It allows you to define the behavior of your application based on the URL requested by the client.

App routes are defined using the @app.route() decorator in Flask. This decorator specifies the URL pattern that should trigger the associated function when a client makes a request to that particular URL.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the 
url_for() function.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Welcome to your profile, {username}!'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'Post ID: {post_id}'

if __name__ == '__main__':
    with app.test_request_context():
        # Generate a URL for the 'index' function
        print(url_for('index'))  # Output: /

        # Generate a URL for the 'user_profile' function with username='John'
        print(url_for('user_profile', username='John'))  # Output: /user/John

        # Generate a URL for the 'show_post' function with post_id=123
        print(url_for('show_post', post_id=123))  # Output: /post/123
```

In Flask, the url_for() function is used for URL building. It generates a URL for a specific function based on its name and any arguments provided.
