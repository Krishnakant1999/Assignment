Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In pandas, a DataFrame is a two-dimensional data structure that represents a tabular, spreadsheet-like data object containing rows and columns. It is a primary data structure used in pandas for data manipulation and analysis.

Series is a one-dimensional labeled array capable of holding any data type. It is essentially a single column of a DataFrame. 

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Emma', 'Michael'],
        'Age': [25, 30, 35],
        'Country': ['USA', 'Canada', 'Australia']}
df = pd.DataFrame(data)

print(df)
```

```python
import pandas as pd

# Creating a Series
data = [4, 8, 15, 16, 23, 42]
series = pd.Series(data)

print(series)
```

Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can 
you give an example of when you might use one of these functions?

head() and tail(): These functions allow you to view the first or last few rows of a DataFrame, respectively.

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Emma', 'Michael'],
        'Age': [25, 30, 35],
        'Country': ['USA', 'Canada', 'Australia']}
df = pd.DataFrame(data)

print(df.head(2))  # Displays the first 2 rows of the DataFrame
print(df.tail(1))  # Displays the last row of the DataFrame
```

describe(): This function generates descriptive statistics of the DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartiles.

```python
import pandas as pd

# Creating a DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50],
        'C': [100, 200, 300, 400, 500]}
df = pd.DataFrame(data)

print(df.describe())
```

shape: This attribute returns the dimensions of the DataFrame as a tuple (rows, columns).


```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Emma', 'Michael'],
        'Age': [25, 30, 35],
        'Country': ['USA', 'Canada', 'Australia']}
df = pd.DataFrame(data)

print(df.shape)  
```

sort_values(): This function allows you to sort the DataFrame by one or more columns.

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Emma', 'Michael'],
        'Age': [25, 30, 35],
        'Country': ['USA', 'Canada', 'Australia']}
df = pd.DataFrame(data)

df_sorted = df.sort_values(by='Age', ascending=False)
print(df_sorted)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel

Series: A Series in pandas is mutable, meaning you can modify its values, add or remove elements, or change its properties.

DataFrame: A DataFrame in pandas is also mutable. You can modify the values in individual cells, add or remove rows or columns, and perform various operations to manipulate the data within the DataFrame.

Panel: Panels were a data structure in older versions of pandas that represented three-dimensional data. However, Panels have been deprecated since pandas version 0.20.0, and they are no longer actively supported. Panels were replaced by other data structures like MultiIndex DataFrames or 3D NumPy arrays. Panels were immutable, meaning they could not be modified after creation.

In summary, Series and DataFrame are mutable, while Panel is immutable (but also deprecated in recent versions of pandas).








Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Creating Series
name_series = pd.Series(['John', 'Emma', 'Michael'])
age_series = pd.Series([25, 30, 35])
country_series = pd.Series(['USA', 'Canada', 'Australia'])

# Combining Series into DataFrame
data = {'Name': name_series, 'Age': age_series, 'Country': country_series}
df = pd.DataFrame(data)

print(df)
```
